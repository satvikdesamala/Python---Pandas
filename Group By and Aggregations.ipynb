```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\kallzz\Desktop\Data Analytics Stuff\Data Analyst - Boot Camp\Python - Jupyter Notebooks\Datasets\Flavors.csv")
df
```

```python
# Group By can be done on a column that has duplicated or repeated values
df.groupby("Base Flavor")
```

```python
# to display the group by data we need to use any applicable method on groupby data
df.groupby("Base Flavor").mean()
```

```python
# Aggregations are performed only on the numeric values
df.groupby("Base Flavor").sum()
```

```python
# min and max can be applied on both numrical and non-numerical. String data will be displayed based on alphabetical order. Min - a-z, Max - z-a
df.groupby("Base Flavor").min()
```

```python
df.groupby("Base Flavor").max()
```

```python
df.groupby("Base Flavor").count()
```

```python
# agg function retunrs specific aggregation values at specific columns
df.groupby("Base Flavor").agg({'Flavor Rating': ['mean', 'max', 'min', 'count']})
```

```python
df.groupby("Base Flavor").agg({'Flavor Rating': ['mean', 'max', 'min', 'count'], 'Texture Rating': ['mean', 'max', 'min', 'count']})
```

```python
# Group by on multiple columns
df.groupby(["Base Flavor", "Liked"]).mean()
```

```python
# describe() can be used to get the stats on numerical data
df.groupby(["Base Flavor", "Liked"]).describe()
```
