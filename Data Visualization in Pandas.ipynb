```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.set_index("Date",inplace=True)
```

```python
df
```

```python
# plot function has a default visualization as Line
df.plot()
```

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

## Horizontal Bar

```python
df.plot.barh(stacked = True)
```

## Scatter Plot

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating')
```

```python
# increase size of dots
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 100)
```

```python
# change color of circles
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 100, c = 'Green')
```
