```python
df2 = df.groupby('Continent').mean().sort_values(by = '2022 Population', ascending = False)
df2
```

```python
# to transpose
df2 = df2.transpose()
```

```python
df.columns
```

```python
df3 = df.groupby('Continent')[['2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']].mean().sort_values('2022 Population', ascending = False)
df3
```

```python
# another way to choose specific columns for analysis
df3 = df.groupby('Continent')[df.columns[5:13]].mean().sort_values('2022 Population', ascending = False)
df3
```

```python
df3.plot()
```

```python
# data plotted from 2022 --> 1970 which is reverse and not recommended to find out the trend. Change the data flow from 1970 --> 2022

df3 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values('2022 Population', ascending = False)
df3
```

```python
df4 = df3.transpose()
```

```python
df4.plot()
```
